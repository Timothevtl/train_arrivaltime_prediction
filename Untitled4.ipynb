```python
import pandas as pd
from sklearn.model_selection import GroupShuffleSplit
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.preprocessing import LabelEncoder

# Load and preprocess the dataset
delays_df = pd.read_csv('D:/s9/machine leanring use cases/delays.csv')
delays_df = delays_df.drop_duplicates()

# Create a unique journey identifier
delays_df['journey_id'] = (
    delays_df['id'] + '_' +
    delays_df['date'] + '_' +
    delays_df['connection'] + '_' +
    delays_df['arrival'] + '_' +
    delays_df['name']
)


# Convert 'delay' to integer and handle missing values
delays_df['delay'] = delays_df['delay'].apply(lambda x: int(''.join(filter(str.isdigit, str(x)))) if pd.notna(x) else None)

# Convert 'arrival' and 'datetime' to datetime, handling errors
delays_df['arrival'] = delays_df['arrival'].replace('Nie dotyczy', pd.NaT)
delays_df['arrival'] = pd.to_datetime(delays_df['arrival'], errors='coerce')
delays_df['datetime'] = pd.to_datetime(delays_df['datetime'])
delays_df['hour'] = delays_df['datetime'].dt.hour
delays_df['day_of_week'] = delays_df['datetime'].dt.dayofweek

# Calculate 'time_to_destination' and filter the dataframe
delays_df['time_to_destination'] = (delays_df['arrival'] - delays_df['datetime']).dt.total_seconds() / 60
delays_df = delays_df[delays_df['time_to_destination'] >= 60]
delays_df['real arrival'] = delays_df.groupby(['journey_id'])['datetime'].transform('last')
delays_df['real delay'] = (delays_df['real arrival'] - delays_df['arrival']).dt.total_seconds() / 60  # Convert to minutes
delays_df['delay gap'] = delays_df['real delay'] - delays_df['delay']

# GroupShuffleSplit based on journey_id...
gss = GroupShuffleSplit(n_splits=1, test_size=0.2, random_state=0)
train_idx, test_idx = next(gss.split(delays_df, groups=delays_df['journey_id']))
train_df = delays_df.iloc[train_idx]
test_df = delays_df.iloc[test_idx]

# Label encoding
label_encoder_carrier = LabelEncoder()
label_encoder_connection = LabelEncoder()

# Combine, Encode, then Split
combined_df = pd.concat([train_df, test_df])
combined_df['carrier_encoded'] = label_encoder_carrier.fit_transform(combined_df['carrier'])
combined_df['connection_encoded'] = label_encoder_connection.fit_transform(combined_df['connection'])

train_df = combined_df.iloc[train_idx]
test_df = combined_df.iloc[test_idx]

# Selecting features and target
features = ['hour', 'day_of_week', 'carrier_encoded', 'connection_encoded']
target = 'real delay'

# Train the model
model = RandomForestRegressor(n_estimators=100, random_state=0)
model.fit(train_df[features], train_df[target])

# Predict and evaluate the model
y_pred = model.predict(test_df[features])
mae = mean_absolute_error(test_df[target], y_pred)
rmse = np.sqrt(mean_squared_error(test_df[target], y_pred))

print(f'Mean Absolute Error: {mae} minutes')
print(f'Root Mean Squared Error: {rmse} minutes')
```

```python

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Assuming df_pos is your DataFrame# Label encoding
label_encoder_carrier = LabelEncoder()
label_encoder_connection = LabelEncoder()
df_pos['carrier_encoded'] = label_encoder_carrier.fit_transform(df_pos['carrier name'])
df_pos['connection_encoded'] = label_encoder_connection.fit_transform(df_pos['departure-arrival trainstation'])
df_pos['delay gap'] = abs(df_pos['real delay'] - df_pos['delay'])
df_pos['time_before_real_arrival'] = (df_pos['real arrival'] - df_pos['Data collection timestamp']).dt.total_seconds() / 60  # Convert to minutes
df_pos['time_before_init_arrival'] = (df_pos['initial arrival'] - df_pos['Data collection timestamp']).dt.total_seconds() / 60  # Convert to minutes
df_pos = df_pos[df_pos['time_before_real_arrival'] >= 120]
# Selecting features and target
features = ['hour', 'day_of_week', 'carrier_encoded', 'connection_encoded']
target = 'real delay'

# Randomly split the dataset into training and testing sets
train_df, test_df = train_test_split(df_pos, test_size=0.2, random_state=0)

# Train the RandomForest model
model = RandomForestRegressor(n_estimators=100, random_state=0)
model.fit(train_df[features], train_df[target])

# Predict and evaluate the model
y_pred = model.predict(test_df[features])
test_df['modelized_delay'] = y_pred  # Add the predictions as a new column

mae = mean_absolute_error(test_df[target], y_pred)
rmse = np.sqrt(mean_squared_error(test_df[target], y_pred))

print(f'Mean Absolute Error: {mae} minutes')
print(f'Root Mean Squared Error: {rmse} minutes')

# Now 'test_df' contains a new column 'modelized_delay' with the predicted delays
```
